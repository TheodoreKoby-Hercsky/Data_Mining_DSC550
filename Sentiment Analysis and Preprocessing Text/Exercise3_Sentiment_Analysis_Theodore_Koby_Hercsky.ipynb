## Part 2: Prepping Text for a Custom Model

## Convert all text to lowercase letters

I will use .str and .lower to convert my review text to lowercase and also create a updated_review column

```python
Train_Data['updated_review'] = Train_Data['review'].str.lower()
```

```python
Train_Data
```

As seen above I have created an updated review column that shows the lowercase text

## Remove punctuation and special characters from the text.

I will use Regex and Lambda function to remove all punctuation and special characters from my updated_review column

```python
import re
Train_Data['updated_review'] = Train_Data['updated_review'].apply(lambda x: re.sub('[^A-Za-z0-9]', ' ', x))
```

```python
Train_Data
```

As seen above the updated_review column has been updated to Remove punctuation and special characters

## Remove stop words.

I will import nltk and download stopwords and then use .apply and lambda to join split and stop my updated_review column

```python
import nltk
nltk.download('stopwords')
stop = stopwords.words('english')
```

```python
Train_Data['updated_review'] = Train_Data['updated_review'].apply(lambda words: ' '.join(word.lower() for word in words.split() if word not in stop))
```

```python
Train_Data
```

As seen above I have gotten rid of the stop words in my update_review column

## Apply NLTK’s PorterStemmer

I will first import porterstemmer and wordnetlemmatizer which will allow me to tokenize your sentences which will split a sentence into words based on the kind of delimiters. Then use apply with my stem_sentences.

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
porter_stemmer = PorterStemmer()
```

```python
def stem_sentences(sentence):
    tokens = sentence.split()
    stemmed_tokens = [porter_stemmer.stem(token) for token in tokens]
    return ' '.join(stemmed_tokens)

Train_Data['updated_review'] = Train_Data['updated_review'].apply(stem_sentences)
```

```python
Train_Data
```
