## Try to calculate the median user score of all the video games. You will likely run into an error because some of the user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. Then, replace all NaN entries in the user score column with the median value.

First I will use the to_numeric approch which will allow me to conver the values in the User_score column into a float. Then using errors='coerce' I will be changing non-numeric values to NaN

```python
Video_Game_Sales['User_Score'] = Video_Game_Sales['User_Score'].apply(pd.to_numeric, errors='coerce')
```

Next I will replace the NaNs with the median value for the 'User_Score' column. This will be done by using replace on the dataframe and using np.nan and setting the User_Score to the median then setting the regex equal to True.

```python
Video_Game_Sales = Video_Game_Sales.replace(np.nan, Video_Game_Sales['User_Score'].median(), regex=True)
```

As seen below by doing this we have converted the User_Score column to a float64 that replaced the NaNs with the median from the column.
