```python
# First I will import some needed libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Load the dataset as a Pandas data frame.

First I will import CSV File heart.csv through the use of Pandas.

```python
heart = pd.read_csv("heart.csv")
heart
```

### Chest Pain & Maximum Heart Rate Achieved by Age Scatter Plot

I will first create a Scatter Plot that shows data from my heart data set that uses the Age attribute to determine if the CP (Chest Pain) and the thalach "Maximum Heart Rate Achieved" attributes from my data set have a similarity between a rise chest pain and Maximum Heart Rate Achieved against the age of the patients.

```python
# x-axis values
x = heart.age
 
#y-axis values
y1 = heart.cp
 
# secondary y-axis values
y2 = heart.thalach
 
# plotting figures by creating aexs object
# using subplots() function
fig, ax = plt.subplots(figsize = (15, 8))
plt.title('Chest Pain & Maximum Heart Rate Achieved by Age')
 
# using the twinx() for creating
# another axes object for secondary y-Axis
ax2 = ax.twinx()
# creating a bar plot
ax.scatter(x, y1, color = 'r')
ax2.scatter(x, y2, color = 'b')

# giving labels to the axises
ax.set_xlabel('Age', color = 'g')
ax.set_ylabel('Chest Pain', color = 'r')
 
# secondary y-axis label
ax2.set_ylabel('Maximum Heart Rate Achieved', color = 'b')
 
# defining display layout
plt.tight_layout()
 
# show plot
plt.show()
```

### Explain what you have learned from your graphs.

In my graph above that is titled "Chest Pain & Maximum Heart Rate Achieved by Age" I have learned that for the most part we are seeing patients with max heart rates in the 120 and above while we are seeing a downword shift as patients are getting older their max heart rate is not as high. And we have more patients that are older that are experiencing higher chest pains. 

### Bar Plot for Resting Blood Pressure & Serum Cholesterol by Age

For my second visualization I am creating a Bar plot that from my heart data set that uses the Age attribute to determine if the trestbps (Resting Blood Pressure) and the chol "Serum Cholesterol" have a similarity between a patients Resting Blood Pressure and Serum Cholesterol against the age of the patients.

```python
# x-axis values
x = heart.age
 
#y-axis values
y1 = heart.trestbps
 
# secondary y-axis values
y2 = heart.chol
 
# plotting figures by creating aexs object
# using subplots() function
fig, ax = plt.subplots(figsize = (15, 8))
plt.title('Resting Blood Pressure Vs. Serum Cholesterol by Age')
 
# using the twinx() for creating
# another axes object for secondary y-Axis
ax2 = ax.twinx()
# creating a bar plot
ax.bar(x, y1, color = 'g')
ax2.bar(x, y2, color = 'b')

# giving labels to the axises
ax.set_xlabel('Age', color = 'r')
ax.set_ylabel('Resting Blood Pressure', color = 'g')
 
# secondary y-axis label
ax2.set_ylabel('Serum Cholesterol', color = 'b')
 
# defining display layout
plt.tight_layout()
 
# show plot
plt.show()
```

### Explain what you have learned from your graphs.

In my graph above that is titled "Resting Blood Pressure Vs. Serum Cholesterol by Age" I have learned that as the patients cholesterol rises or dips so does the patients resting blood preasure. Which we see as the individuals age rises to 70 we see an increase then a drop after the patient passes 70. I believe this visualization tells us as one atribute of an individuals health worsens so does others which causes heart issues and other medical problems.

### Line Plot for Chest Pain & ST Depression Induced by Heart Issue

For my last visualization I am creating a Line plot that from my heart data set that uses the thal attribute that states that 1 = normal; 2 = fixed defect; 3 = reversable defect hearts to determine if these patients have chest pain and if they have ST Depression Induced by exercise that is relative to rest.

```python
# Define Data

x = heart.thal
data_1 = heart.cp
data_2 = heart.oldpeak
  
# Create Plot

fig, ax1 = plt.subplots(figsize = (15, 8)) 
  
ax1.set_xlabel('Heart Issue') 
ax1.set_ylabel('Chest Pain', color = 'red') 
plot_1 = ax1.plot(x, data_1, color = 'red') 
ax1.tick_params(axis ='y', labelcolor = 'red') 
plt.title('Chest Pain & ST Depression Induced by Heart Issue')

# Adding Twin Axes

ax2 = ax1.twinx() 
  
ax2.set_ylabel('ST Depression Induced by Exercise', color = 'green') 
plot_2 = ax2.plot(x, data_2, color = 'green') 
ax2.tick_params(axis ='y', labelcolor = 'green') 

# defining display layout
plt.tight_layout()


# Show plot

plt.show()
```

### Explain what you have learned from your graph.

From my line plot above I learned that for the most part patients that are either 2 = have a fixed heart defect or 3 = have a reversable heart defect are more likely to have chest pain and ST Depression Induced by exercise that is relative to rest which is shown in the visualization. As it seems as if these patients would me vulnerable and more likely to see issues such as chest pain and depression.
